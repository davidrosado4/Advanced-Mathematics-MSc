Now, let us plot the persistence diagram. Although the statement only asks for the study of $H_1$, we are going to study  $H_0,H_1$ and $H_2$, since with the same effort, we complete the exercise even further.

```python
points=np.reshape(points,((50,3)))
diagrams = ripser(points,maxdim=2)['dgms']
plot_diagrams(diagrams, show=True)
```

Finally, let us draw the landscape functions, as said before, to complete even more the exercise, we give the landscape functions for $H_0,H_1$ and $H_2$.

```python
plot_landscape_simple(PersLandscapeExact(diagrams, hom_deg=0),
                             title="Degree 0 Persistence Landscape of the point cloud")
```
